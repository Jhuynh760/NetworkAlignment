```python
import pandas as pd
import matplotlib.pyplot
import numpy as np
matplotlib.pyplot.rcParams["patch.force_edgecolor"] = True
```

```python
n10df = pd.read_csv("n10.csv")
maxDegree = n10df["Degree"].max()
matplotlib.pyplot.xticks(range(0, maxDegree+1))
matplotlib.pyplot.xlim([-1, maxDegree+1])
matplotlib.pyplot.title("n10.txt Histogram")
n10df["Degree"].hist(figsize=(10,10), bins = np.arange(maxDegree+2) - .5, grid = False, linewidth=1.2)
```
